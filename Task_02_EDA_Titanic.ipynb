```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.hist(df['Age'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival vs Gender")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival vs Passenger Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

After performing data cleaning and exploratory data analysis on the Titanic dataset, several important patterns and relationships were identified.

🔎 Data Cleaning Summary

Missing values in Age were filled using the median.

Missing values in Embarked were filled using the mode.

The Cabin column was removed due to a large number of missing values.

Data was checked for duplicates and data types were verified.

Key Findings

1️⃣ Gender Impact on Survival

Female passengers had a significantly higher survival rate compared to male passengers.

This indicates that women were given priority during rescue operations.

2️⃣ Passenger Class Influence

First-class passengers had the highest survival rate.

Third-class passengers had the lowest survival rate.

This shows socioeconomic status played a role in survival.

3️⃣ Age Factor

Younger passengers had better survival chances compared to older passengers.

Children were prioritized during evacuation.

4️⃣ Fare and Survival

Passengers who paid higher fares had higher survival probability.

Fare is positively correlated with survival.

5️⃣ Correlation Insights

Passenger class is negatively correlated with survival (higher class → better survival).

Gender and class were the strongest factors affecting survival.
